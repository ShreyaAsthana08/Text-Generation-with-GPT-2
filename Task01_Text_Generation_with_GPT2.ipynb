```python
from transformers import GPT2Tokenizer , GPT2LMHeadModel
```

# **1- Convert the sentances into tokens**

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
```

```python
model = GPT2LMHeadModel.from_pretrained('gpt2-large' , pad_token_id= tokenizer.eos_token_id)
```

```python
tokenizer.eos_token_id
```

```python
sentance = 'I love painting the nature scenery'
numeric_ids = tokenizer.encode(sentance , return_tensors= 'pt')
```

```python
numeric_ids
```

```python
tokenizer.decode(numeric_ids[0][1])
```

# **2- Generate the text from given sentance**

```python
result = model.generate(numeric_ids , max_length= 100 , num_beams=5, no_repeat_ngram_size =2, early_stopping= True)
```

```python
result
```

```python
generate_text = tokenizer.decode(result[0], skip_special_tokens=True)
print(generate_text)
```
